# Wan2GP Colab Demo

This notebook runs a short-lived Wan2GP (or Gradio) server on a free Google Colab GPU instance and exposes it with `localtunnel` so you can point the desktop app or dev instance at it for testing.

Limitations: sessions are ephemeral (~12 hours) and tunnels are unreliable for production. See `DEPLOY_COLAB.md` for full details.

```python
%%bash
set -euo pipefail
echo 'Installing system deps...'
apt-get update -y && apt-get install -y git wget python3-pip ffmpeg nodejs npm
python3 -m pip install --upgrade pip
echo 'Cloning Wan2GP (if it already exists this is fine)...'
git clone https://github.com/deepbeepmeep/Wan2GP.git || true
cd Wan2GP || exit 1
echo 'Attempting to install Python requirements (may fail if the repo changes)...'
pip install -r requirements.txt || true
echo 'If you have model weights, place them in /content/Wan2GP/models or mount Google Drive.'
echo 'Starting Wan2GP server in background (writes logs to /content/Wan2GP/wgp.log)...'
nohup python wgp.py --listen --server-name 0.0.0.0 --port 7860 > wgp.log 2>&1 &
echo 'Installing localtunnel...'
npm install -g localtunnel
echo ''
echo 'Start a localtunnel to expose port 7860. Pick a subdomain (or omit to get random). Example uses: my-wan2gp-demo'
echo 'Run this in a separate cell to keep the notebook interactive:'
echo 
7860
```
